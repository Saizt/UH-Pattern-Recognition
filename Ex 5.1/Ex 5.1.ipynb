```python
dataset = datasets.MNIST(root="dataset/", download=True)
dataset = dataset.data.numpy().reshape(-1,784).T
```

# Statements (a-d)

```python
u, s, vT = np.linalg.svd(dataset, full_matrices=False)
```

```python
plt.figure(figsize=(15,6))
plt.plot(s)
plt.title('Singulare Value Spectrum')

f, ax = plt.subplots(1,2,figsize=(15,6))
ax[0].imshow(dataset[:,0].reshape(28,28))
ax[0].set_title('The real number image')
ax[1].imshow((u[:,:100]@np.diag(s)[:100,:]@vT[:,:100])[:,0].reshape(28,28))
ax[1].set_title('Reconstructed by 100 modes number image');
```

Seems like 100 modes is enough to reconstruct images, hence the rank of digit space is around this number.

Interpretations of U, $\Sigma$ and V matricies of SVD decomposition:

- U is modes (like patterns)
- $\Sigma$ is singular values
- V is loading scores for the modes (like weights of the corresponding modes/patterns)
